```python
import pandas as pd
import numpy as np
import pickle
```

### Imputation

```python
data = pd.read_pickle('credit_data_with_nans.pickle')
```

```python
data.head()
```

Fill with zeros

```python
zeros = data.fillna(0)
zeros.head()
```

Fill with false

```python
false = data.fillna(False)
false.head()
```

fill with median or mode

```python
median = data.fillna(data.median())
median.head()
```

Fill Nans with interpolated values.

```python
interpolated = data.interpolate()
interpolated.head()
```

```python
interpolated.isna().sum()
```

Fill with preceding values

```python
preceding = data.fillna(method= 'bfill')
preceding.head()
```

Fill NaNs with following values

```python
following = data.fillna(method='ffill')
following.head()
```

Fill with randomly sampled values:

```python
random = data.fillna(data.sample(1))
random.head()
```

Clearly none of these methods make sense on a table-wide basis. Below we try filling Marriage column NaNs with  values

```python
values = data.MARRIAGE.unique()[np.logical_not(np.isnan(data.MARRIAGE.unique()))]
data.MARRIAGE = data.MARRIAGE.fillna(np.random.choice(values))
```

```python
data.MARRIAGE.fillna(data.MARRIAGE.sample(len(data)), inplace=True)
```

```python
data.MARRIAGE.value_counts()
```

```python
data.MARRIAGE.isna().sum()
```
