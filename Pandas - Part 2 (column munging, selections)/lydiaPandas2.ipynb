### Rearranging the dataFrame for ease of use:
Extract list of columns, reaarnge, construct a new df with custom order.

```python
columns = data.columns.to_list()
columns
```

```python
columns.sort()
columns
```

```python
columns.insert(0, columns.pop(columns.index('color')))
```

```python
columns
```

```python
rearranged_data = data[columns]
rearranged_data.head()
```
